```python
page = open('Sternprofs', 'r')
code = page.read()
code
```

```python
regex = re.compile(r'/ShowRatings.jsp\?tid=\d+&amp;showMyProfs=true')
matches = regex.finditer(code)
idlist = []
for match in matches:
    idlist += ' http://www.ratemyprofessors.com' + match.group()
new_id_list = ''.join(idlist)
finallist = new_id_list.split(' ')
del finallist[0]
```

```python
lastNameList = []
for link in finallist:
    response = requests.get(link)
    doc = html.fromstring(response.text)
    try:
        profLastName = doc.xpath('//span[@class="plname"]')[0].text_content()
        profLastName = profLastName.strip()
    except:
        profLastName = "N/A"
    lastNameList.append(profLastName)
lastNameList
```

```python
finalList = []
for link in finallist:
    response = requests.get(link)
    doc = html.fromstring(response.text)
    #rmprating
    try:
        rmpRating = doc.xpath('//div[@class="grade"]')[0].text_content()
        rmpRating = rmpRating.strip()
    except:
        rmpRating = ''
    #rmdifficulty
    try:
        rmpDiff = doc.xpath('//div[@class="grade"]')[2].text_content()
        rmpDiff = rmpDiff.strip()
    except:
        rmpRating = ''
    #First Name
    try:
        profFirstName = doc.xpath('//span[@class="pfname"]')[0].text_content()
        profFirstName = profFirstName.strip()
    except:
        profFirstName = ""
    #Last Name
    try:
        profLastName = doc.xpath('//span[@class="plname"]')[0].text_content()
        profLastName = profLastName.strip()
    except:
        profLastName = ""
    #Comments
    try:
        #1
        comments1 = doc.xpath('//p[@class="commentsParagraph"]')[0].text_content()
        comments1 = comments1.strip()
        comments1rating = doc.xpath('//span[@class="score good"]')[0].text_content()
        comments1difficulty = doc.xpath('//span[@class="score inverse average"]')[0].text_content()
    except:
        comments1 = ''
        comments1rating = ''
        comments1difficulty = ''
    #2
    try:
        comments2 = doc.xpath('//p[@class="commentsParagraph"]')[1].text_content()
        comments2 = comments2.strip()
        comments2rating = doc.xpath('//span[@class="score good"]')[1].text_content()
        comments2difficulty = doc.xpath('//span[@class="score inverse average"]')[1].text_content()
    except:
        comments2 = ''
        comments2rating = ''
        comments2difficulty = ''
    #3
    try:
        comments3 = doc.xpath('//p[@class="commentsParagraph"]')[2].text_content()
        comments3 = comments3.strip()
        comments3rating = doc.xpath('//span[@class="score good"]')[2].text_content()
        comments3difficulty = doc.xpath('//span[@class="score inverse average"]')[2].text_content()
    except:
        comments3 = ''
        comments3rating = ''
        comments3difficulty = ''
    #tags
    try:
        tag1 = doc.xpath('//span[@class="tag-box-choosetags"]')[0].text_content()
        tag1 = tag1.strip()
    except:
        tag1 = ''
    try:
        tag2 = doc.xpath('//span[@class="tag-box-choosetags"]')[1].text_content()
        tag2 = tag2.strip()
    except:
        tag2 = ''
    try:
        tag3 = doc.xpath('//span[@class="tag-box-choosetags"]')[2].text_content()
        tag3 = tag3.strip()
    except:
        tag3 = ''
    
    entry = {
            'First Name' : profFirstName,
            'Last Name' : profLastName,
            'Tag 1': tag1,
            'Tag 2': tag2,
            'Tag 3': tag3,
            'Comment 1' : comments1,
            'Comment 1 Rating': comments1rating,
            'Comment 1 Difficulty': comments1difficulty,
            'Comment 2' : comments2,
            'Comment 2 Rating': comments2rating,
            'Comment 2 Difficulty': comments2difficulty,
            'Comment 3' : comments3,
            'Comment 3 Rating': comments3rating,
            'Comment 3 Difficulty': comments3difficulty,
            'Rate My Prof Rating': rmpRating,
            'Rate My Prof Difficulty': rmpDiff
        }
    finalList.append(entry)
counter = 0
for entity in finalList:
    if entity['First Name'] == "" or entity['Last Name'] == '' or entity['Rate My Prof Rating'] == '':
        del finalList[counter]
    counter += 1
```
