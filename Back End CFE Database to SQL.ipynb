```python
def getRating(url):
    doc = html.fromstring(requests.get(url).text)
    page = doc.xpath('//div[@id="mainContent"]')
    
    result = []
    for i in page:
        fname = i.xpath('.//span[@class="pfname"]')[0].text_content().strip()
        lname = i.xpath('.//span[@class="plname"]')[0].text_content().strip()
        rating = i.xpath('.//div[@class="grade"]')[0].text_content().strip()
        levelOfDifficulty = i.xpath('.//div[@class="breakdown-section difficulty"]/div[@class="grade"]')[0].text_content().strip()
        
        entry = {
            "first name" : fname,
            "last name" : lname,
            "rating" : rating,
            "level of difficulty" : levelOfDifficulty,
        }
        result.append(entry)
    return result
```
